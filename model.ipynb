## Target Variable: ev_market_share

### Histogram with KDE (Kernel Density Estimation)

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=data['ev_market_share'], kde=True)
plt.show()
```

This histogram provides a clear "snapshot" of the current state of global EV adoption in the dataset. Here is the summary of what it tells us:

1. Most countries have "Non-Adoption"

The tallest bar (at the far left) shows that nearly 700 of the 1,200 rows (~58%) have an EV market share close to 0–2%.

    Conclusion: Despite the headlines, for most countries and vehicle segments in the historical data, EVs have barely scratched the surface.

2. Heavy Right Skew

The data has a "long tail" that stretches all the way to 100%.

    Conclusion: Statistically, the target variable is not normally distributed. This is a "power-law" or exponential-looking distribution where a few "leaders" (like Norway) are miles ahead of everyone else.

3. The "Tipping Point" Gap

Notice how the bars drop off sharply after the 10% mark.

    Conclusion: Very few regions in the data have reached a "steady-state" middle ground (e.g., 30-50% share). It seems countries are either at the very beginning (0-10%) or they are the extreme outliers (80%+).

### Boxplot

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=data['ev_market_share'])
plt.show()
```

This Box Plot is the best tool for spotting Outliers and understanding the "typical" experience versus the "exceptions" in the data.

Here is the summary of what it reveals:

1. The Median is Nearly Zero

The vertical line inside the blue box is very close to the left axis.

    Conclusion: More than 50% of the rows in your dataset have an EV market share of less than 1%. For most of the world (historically), EVs aren't even a factor yet.
2. The "Typical" Range (The Box)

The blue box represents the middle 50% of your data (The Interquartile Range).

    Conclusion: Most of the data lives between 0% and ~6%. This is the "comfort zone" where the machine learning model will have the most examples to learn from.
3. The "Leading Edge" (The Right Whisker)

The vertical line to the right of the box (around the 15% mark) shows the limit of what is considered "statistically normal" for this dataset.

    Conclusion: Any country with more than ~15% market share is officially an exception to the rule.
4. The Outliers (The Dots)

Every single circle we see past 15% is a specific data point that "broke the mold."

    Conclusion:
    The Cluster (20% - 40%): These are likely modern-day China, Germany, and the UK. They are the "early adopters" who have successfully moved past the tipping point.
    The Extremes (80%+): These are the global leaders, specifically Norway in the last few years.

### Time Series Lineplot based on region

```python
plt.figure(figsize=(6, 4))
sns.lineplot(data=data, x='year', y='ev_market_share', hue="region")
```

This Time Series Plot shows the speed of EV adoption across different parts of the world. Here are the key takeaways for the graph:

1. Europe is the Clear Leader (and most diverse)

The orange line representing Europe isn't just the highest; it has the steepest slope.

    The Shaded Area: Notice how wide the orange shading is. This reflects a massive "performance gap" within Europe (e.g., Norway's 90% share vs. Poland's 5% share).

    Insight: "Region = Europe" will be one of the strongest predictors, but it also has the most uncertainty.

2. The 2020 "Global Pivot"

Almost every line on this chart (except South America) shows a distinct "elbow" or tipping point between 2018 and 2020.

    Observation: Before 2018, growth was slow and flat. After 2020, it became exponential.

3. The "Follower" Pack

APAC (purple), Oceania (blue), and North America (red) are currently in a tight battle.

    They are roughly 5–7 years "behind" Europe's trajectory but are following the exact same exponential curve shape.

    Trend: They are just now hitting the 10% market share mark that Europe hit around 2017.

4. South America's Lag

The green line (South America) is the only region that has stayed relatively flat, only recently crossing the 1–3% mark.

    Observation: This suggests that economic factors (like GDP) or different infrastructure priorities in this region are heavily resisting the global EV trend.

### Correlation Matrix




```python
import numpy as np

# 1. Calculate correlation
corr = data.select_dtypes(include=['number']).corr()

# 2. Create a mask to hide the redundant top half
# mask = np.triu(np.ones_like(corr, dtype=bool))

# 3. Set a much larger figure size
plt.figure(figsize=(16, 12))

# 4. Plot with fmt=".2f" (limits decimals) and smaller font size
sns.heatmap(corr, 
            # mask=mask, 
            annot=True, 
            fmt=".2f",      # Rounds to 2 decimal places
            cmap='coolwarm', 
            annot_kws={"size": 9},  # Reduces font size of numbers
            cbar_kws={"shrink": .8})

plt.title('Correlation Heatmap (Optimized)', fontsize=16)
plt.xticks(rotation=45, ha='right') # Tilts labels to prevent overlap
plt.show()
```

### Show High Correlation Features

```python
MULTICOL_THRESHOLD = 0.7
# 1. Convert to 1D Series and remove self-correlations (1.0)
corr_series = corr.unstack()
high_corr = corr_series[abs(corr_series) > MULTICOL_THRESHOLD]
high_corr = high_corr[high_corr < 1.0]      # Filter out the 1.0 diagonal

# 2. Display unique pairs (to avoid seeing A-B and B-A)
print("High Correlation Pairs (> 0.7):")
print(high_corr.sort_values(ascending=False).drop_duplicates())
```

### Show High Correlation Features with Target (ev_market_share)

```python
# 1. Define target and threshold
target = 'ev_market_share'
threshold = 0.5  # Adjust this (e.g., 0.3 for more features, 0.7 for fewer)

# 2. Get correlations with the target specifically
target_corr = data.select_dtypes(include=['number']).corr()[target]

# 3. Filter for features higher than the threshold (absolute value)
# We exclude the target itself (where correlation is 1.0)
high_target_corr = target_corr[abs(target_corr) > threshold].drop(labels=[target])

# 4. Display result
print(f"Features with absolute correlation > {threshold} with {target}:")
print(high_target_corr.sort_values(ascending=False))
```
